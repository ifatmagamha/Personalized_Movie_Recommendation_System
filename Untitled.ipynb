query= """
CREATE OR REPLACE TABLE `students-group1.movie_reco.movie_stats` AS
SELECT
  movieId,
  COUNT(*) AS n_ratings,
  AVG(rating) AS avg_rating
FROM `master-ai-cloud.MoviePlatform.ratings`
GROUP BY movieId;
"""   
job = client.query(query)
job.result()
