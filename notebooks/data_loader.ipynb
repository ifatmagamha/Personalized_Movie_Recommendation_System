```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent
DATA_DIR = PROJECT_ROOT / "data"

sys.path.append(str(PROJECT_ROOT))
```

```python
from google.cloud import bigquery
from src.preprocessing import (
    get_interactions,
    get_movie_stats,
    get_user_stats
)

client = bigquery.Client()

interactions = get_interactions(
    client,
    min_user_ratings=5,
    min_movie_ratings=20,
    limit=None
)

movie_stats = get_movie_stats(client, min_ratings=20, limit=None)
user_stats = get_user_stats(client, min_ratings=5, limit=None)

interactions.to_parquet("data/interactions.parquet")
movie_stats.to_parquet("data/movie_stats.parquet")
user_stats.to_parquet("data/user_stats.parquet")
```

```python
from google.cloud import bigquery
from src.preprocessing import get_movies_metadata
from pathlib import Path

client = bigquery.Client()
movies = get_movies_metadata(client, limit=None)

Path("data").mkdir(exist_ok=True)
movies.to_parquet("data/movies.parquet", index=False)
print(movies.shape)
```
